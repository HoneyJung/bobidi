```python
import nltk
f = open('kinglear.txt', mode='rt', encoding='utf-8')
text = ''
for a in f.read():
    if a.isalpha() or a=='\t' or a == ' ' or a=='\n' or a=='.':
        text = text + a
#     if a==';' or a == ':':
#         text = text + '.'
        
text = text.lower().replace('\n\n', '\n').replace('\t\t', '\t')
sentences = text.split('.')
for sentence in sentences:
    print(sentence)
```

# 1. Uni_gram

```python
from collections import defaultdict
uni_gram = defaultdict(int)
for sentence in sentences:
    for word in sentence.split():
        uni_gram[word] = uni_gram[word] + 1

sorted(uni_gram.items(), key = lambda item: item[1], reverse = True)[:10]
    
```

# 2. Bi_gram

```python
bi_gram = defaultdict(int)
for sentence in sentences:
    words = sentence.split()
    for i in range(len(words)):
        if i == 0:
            before = words[i]
        else:
            bi_gram[(before, words[i])] = bi_gram[(before, words[i])] + 1
            before = words[i]

sorted(bi_gram.items(), key = lambda item: item[1], reverse = True)[:10]
```
